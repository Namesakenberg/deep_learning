```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from keras.preprocessing import image   # helps working with images
from tensorflow.keras.preprocessing.image import ImageDataGenerator  # use to generate images during augumentation
```

```python
img = image.load_img('/content/Cat.jpg',target_size=(200,200))    # compress the size of image to 200,200
```

```python
plt.imshow(img)
```

```python
print(type(img))   # not a numpy array , the image is in the form of PIL img
```

# generate new images from the exisiting ones

```python
datagen = ImageDataGenerator(
    rotation_range =  30 ,       # angle of rotation
    shear_range = 0.2,
    zoom_range = 0.2,            # zooming the image
    horizontal_flip = True,      # mirror image
    width_shift_range = 0.2,     # shifting the width (moving left/right)
    height_shift_range = 0.2,    # shifting the height (moving up/down)
)
```

```python
img = image.img_to_array(img)     # convert PIL image to numpy array
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)    # 200,200,3 = size of image
                                          # 1 = batch_size / no. of images being processed
```

```python
i=0

for output in datagen.flow(input_batch , batch_size=1 , save_to_dir='Aug_images'):
  i+=1
  if i==10:
    break


# make a directory 'Aug_images' to save the augumented image
# run a loop from 1 to 10 to get 10 augumented images

# datagen.flow is used to process a single image
# datagen.flow_from_directory is used to process a complete batch of images
```
